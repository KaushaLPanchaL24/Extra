```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataaa.csv')
data
```

```python
x_data = data['x']
y_data = data['y']
```

```python
n = len(x_data)
n
```

```python
x_bar = x_data.sum()/n
x_bar
```

```python
y_bar = y_data.sum()/n
y_bar
```

```python
numer = np.sum((x_data - x_bar) * (y_data - y_bar))
denom = np.sum((x_data - x_bar) ** 2)
b1 = (numer/denom)
b1
```

```python
b0 = y_bar - (b1*x_bar)
b0
```

```python
y_cab = b0 + (b1*x_data)
y_cab
```

```python
SST = np.sum((y_data - y_bar) ** 2)
SST
```

```python
SSE = np.sum((y_data - y_cab)**2)
SSE
```

```python
SSR = SST - SSE
SSR
```

```python
MSE = SSE/(n-2)
MSE
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, color="maroon", s=80, label="observed data")
plt.plot(x_data, y_cab, color="green", linewidth=2, label=f'Best fit line Y = {b0:.2f} + {b1:.2f}x')
plt.axhline(y_bar, color="blue", linestyle='--', label=f'Y = {y_bar:.2f}')

plt.title("Scatter diagram with best fit line...")
plt.xlabel("X values", fontsize=12)
plt.ylabel("Y values", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
residual = y_data - y_cab   
print("Residual (Actual - Predicted) : ")
print(residual)
```

```python
df_results = pd.DataFrame({
    'X' : x_data,
    'Actual Y' : y_data,
    'Predicted Y' : y_cab,
    'Residual' : residual
})

df_results
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_data, residual, color='purple',label='Error point')
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("X values")
plt.ylabel("Residuals")
plt.legend()
plt.grid(True)
plt.show()
```
